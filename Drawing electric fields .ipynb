## Drawing Electric fields with python
based on code from geeksforgeeks https://www.geeksforgeeks.org/matplotlib-pyplot-streamplot-in-python/

## Run the next cell as setup:

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

 
# Function to determine electric field
def E(q, r0, x, y):
    den = np.hypot(x-r0[0], y-r0[1])**3
    return q * (x - r0[0]) / den, q * (y - r0[1]) / den


# Grid of x, y points
nx, ny = 1000,1000
x = np.linspace(-3, 3, nx)
y = np.linspace(-3, 3, ny)
X, Y = np.meshgrid(x, y)
 
def plotFields(charges):
    print("Charges:" , charges)
    # Electric field vector, E =(Ex, Ey)
    # as separate components
    Ex, Ey = np.zeros((ny, nx)), np.zeros((ny, nx))
    
    for charge in charges:
        ex, ey = E(*charge, x = X, y = Y)
        Ex += ex
        Ey += ey
    
    fig = plt.figure(figsize =(18, 8))
    ax = fig.add_subplot(111)
    
    # Plotting the streamlines with
    # proper color and arrow
    color = 2 * np.log(np.hypot(Ex, Ey))

    ax.streamplot(x, y, Ex, Ey,
                linewidth = 0.75, cmap = plt.cm.inferno,
                density = 5, arrowstyle ='->',
                arrowsize = 0.5)
    
    # Add filled circles for the charges
    # themselves
    charge_colors = {True: '#AA0000',
                    False: '#0000AA'}
 
    for q, pos in charges:
        ax.add_artist(Circle(pos, 0.05, color = charge_colors[q>0]))
    
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal')
    
    plt.show()
```

## Example 1: Dipole with opposite charges
Define the charges here using the format: 


``charges = [(q1, (x1, y1)), (q2, (x2, y2)), (q3, (x3, y3)), etc]``

```python
charges = [(-1, (1.0, 0.0)), (1, (-1.0, 0))]
plotFields(charges)
```

## Example 2: Dipole with equal charges

```python
charges = [(1, (1.0, 0.0)), (1, (-1.0, 0))]
plotFields(charges)
```

## Example 3: Dipole with very different charges

```python
charges = [(-10, (1.5, -1.5)), (1, (-1.5, 0))]
plotFields(charges)
```

## Example 4: Square

```python
charges = [(-1, (1.0, 0.0)), (1, (-1.0, 0)),(-1, (0.0, -1.0)), (1, (0.0, 1))]
plotFields(charges)
```

## Example 5: Square 2

```python
charges = [(-1, (1.5, 1.5)), (5, (-1.5, -1.5)),(-3, (0, -1.0)), (1, (1, 1))]
plotFields(charges)
```
